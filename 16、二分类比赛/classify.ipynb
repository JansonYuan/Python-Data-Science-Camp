## 1.LogisticRegression 0.71892862

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(train_X,train_y)
```

```python
#测试
test_y=LR.predict(test_data)
test_y
```

```python
df_test=pd.read_csv('test_set.csv')
df_test['pred']=test_y.tolist()
df_result=df_test.loc[:,['ID','pred']]
df_result.to_csv('LR.csv',index=False)
```

## 3.SVM---0.71738981

```python
from sklearn.svm import LinearSVC
classifierSVM=LinearSVC()
classifierSVM.fit(train_X,train_y)
```

```python
test_ySVM=classifierSVM.predict(test_data)
df_test=pd.read_csv('test_set.csv')
df_test['pred']=test_ySVM.tolist()
df_result=df_test.loc[:,['ID','pred']]
df_result.to_csv('LSVM.csv',index=False)
```

## 4.KNN---0.60200031

```python
from sklearn.neighbors import KNeighborsClassifier
classifierKNN=KNeighborsClassifier()
classifierKNN.fit(train_X,train_y)
```

```python
test_yKNN=classifierKNN.predict(test_data)
df_test=pd.read_csv('test_set.csv')
df_test['pred']=test_yKNN.tolist()
df_result=df_test.loc[:,['ID','pred']]
df_result.to_csv('KNN.csv',index=False)
```

## 5.决策树  0.65560590

```python
from sklearn.tree import DecisionTreeClassifier
classifierTree=DecisionTreeClassifier()
classifierTree.fit(train_X,train_y)
```

```python
test_yTree=classifierTree.predict(test_data)
df_test=pd.read_csv('test_set.csv')
df_test['pred']=test_yTree.tolist()
df_result=df_test.loc[:,['ID','pred']]
df_result.to_csv('tree.csv',index=False)
```
